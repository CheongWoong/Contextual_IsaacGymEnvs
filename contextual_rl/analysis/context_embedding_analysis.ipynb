```python
import torch
```

```python
seed = 0
```

```python
train_env_name = 'ContextualAntTrain_osi_true'
# train_env_name = 'ContextualAntTrain_osi_true_student'

# TODO: extract all test envs for PCA computation and visualization
env_name = 'ContextualAntTestEasy1'

filename = f'../runs/test/seed_{seed}/{env_name}/{train_env_name}/99942400/contexts.pt'
# filename = f'../runs/test/seed_{seed}/{env_name}/{train_env_name}/99942400_phase2/contexts.pt'
# filename = f'../runs/test/seed_{seed}/{env_name}/{train_env_name}/99942400_phase2_raw/contexts.pt'

contexts = torch.load(filename)
```

```python
contexts = torch.cat(contexts, dim=0)
```

```python
contexts.shape
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
embeddings = contexts
```

```python
pca = PCA(n_components=2)
embeddings_2d_pca = pca.fit_transform(embeddings)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(embeddings_2d_pca[:, 0], embeddings_2d_pca[:, 1])
plt.title('PCA-reduced Embeddings')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()
```
